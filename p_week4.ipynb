### Concatenate DataFrame - attach DataFrame or Series together

```python
data = np.zeros( len(df))   ## create a numpy array of zeros
index = np.array(range(1, len(df)+1))   ## create an array of numbers from 1 to length of dataframe
new_ser= pd.Series(data, name = 'zeros', index=index) ## create a series with name zeros and index as index
new_df = pd.concat(objs=[df, new_ser], axis=1) ## concatenate the series to the dataframe

new_df.head (5) ## create
# axis{0/â€™indexâ€™, 1/â€™columnsâ€™}, default 0
```
